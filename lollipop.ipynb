```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/avalanches.csv')

n = len(df)
print(n)

df.head()
```

```python
plt.bar(df.Year, df.Fatalities)
plt.show()
```

```python
fig, ax = plt.subplots(1, figsize=(16,8))
colors = ['black'] + ((n-1)*['lightgrey'])

plt.bar(df.Year, df.Fatalities, color=colors)
plt.scatter(df.Year, df.Fatalities, color=colors)
plt.show()
```

```python
fig, ax = plt.subplots(1, figsize=(16,8))
colors = ['black'] + ((n-1)*['lightgrey'])

plt.bar(df.Year, df.Fatalities, color=colors, width=0.2)
plt.scatter(df.Year, df.Fatalities, color=colors, s=100)

plt.show()
```

```python
fig, ax = plt.subplots(1, figsize=(24,8))
colors = ['black'] + ((n-1)*['lightgrey'])

plt.bar(df.Year, df.Fatalities, width=0.3, color=colors)
plt.scatter(df.Year, df.Fatalities, color=colors, s=100)

# remove spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
    
plt.ylim(0,)
plt.xlim(1950,2022)

plt.title('US Avalanche Fatalities by Year', loc='left', fontsize=16)
```

```python
fig, ax = plt.subplots(1, figsize=(12,8))
plt.scatter(df.Year, df.Fatalities)
```

```python
fig, ax = plt.subplots(1, figsize=(12,8))
colors = ['black'] + ((n-1)*['lightgrey'])

for idx, val in df.iterrows():
    plt.plot([val.Year, val.Year], [-1, val.Fatalities], color=colors[idx])

plt.show()
```

```python
fig, ax = plt.subplots(1, figsize=(12,8))
colors = ['black'] + ((n-1)*['lightgrey'])

for idx, val in df.iterrows():
    plt.plot([val.Year, val.Year], [-1, val.Fatalities], color=colors[idx], marker='o')

plt.ylim(0,)
plt.show()
```

```python
fig, ax = plt.subplots(1, figsize=(12,8))
color = 'b'
colors = ['black'] + ((len(df)-1)*['lightgrey'])

for idx, val in df.iterrows():
    plt.plot([val.Year, val.Year], [-1, val.Fatalities], color='black', marker='o', lw=4, markersize=10)
    plt.plot([val.Year, val.Year], [-1, val.Fatalities], color=colors[idx], marker='o', markersize=8)

# remove spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
    
plt.ylim(0,)
plt.xlim(1950,2022)

plt.title('US Avalanche Fatalities by Year', loc='left', fontsize=16)
plt.text(2021, -3, 'Source:https://avalanche.state.co.us/accidents/statistics-and-reporting/', ha='right')

fatalities_2021 = df[df['Year'] == 2021].Fatalities.values[0]
plt.text(2021, fatalities_2021+1, fatalities_2021, ha='center')

plt.savefig('chart.png')
```

```python
fig, ax = plt.subplots(1, figsize=(12,8))
color = 'b'
colors = ['#E74C3C'] + ((len(df)-1)*['#F5B7B1'])

for idx, val in df.iterrows():
    plt.plot([val.Year, val.Year], [-1, val.Fatalities], color=colors[idx], marker='o', lw=4, markersize=10, markerfacecolor='#E74C3C')

# remove spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
    
plt.ylim(0,)
plt.xlim(1950,2022)

plt.title('US Avalanche Fatalities by Year', loc='left', fontsize=16)
plt.text(2021, -3, 'Source:https://avalanche.state.co.us/accidents/statistics-and-reporting/', ha='right')

fatalities_2021 = df[df['Year'] == 2021].Fatalities.values[0]
plt.text(2021, fatalities_2021+1, fatalities_2021, ha='center')

plt.savefig('chart.png')
```
